```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('Online Sales Data Task 1.csv')
df.info()
```

```python
df.head()
```

```python
# Checking null values
df.isnull().sum()
```

# There are no Null values in this data set

```python
#checking dupliate rows
df.duplicated().sum()
```

# There are no dulicated rows or data in our data set

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()           # Remove leading/trailing spaces
    df[col] = df[col].str.lower()                        # Convert to lowercase
    df[col] = df[col].str.title()
df.head()
```

```python
df.tail(20)
```

```python
#Changing the datatype of the Date colum
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'].dtype


```

# we have converted the data type of the "Date" column to datetime from object

```python
df.columns
```

```python
df.rename(columns={
    'Transaction ID': 'TransactionID',
    'Product Category': 'ProductCategory',
    'Product Name': 'ProductName',
    'Units Sold': 'UnitsSold',
    'Unit Price': 'UnitPrice',
    'Total Revenue': 'TotalRevenue',
    'Payment Method': 'PaymentMethod'
}, inplace=True)

df.columns
```

# We have removed the speces between the columns

The dataset was thoroughly cleaned and prepared for analysis. There were no missing values or duplicate rows found in the data. The "Date" column was successfully converted from object to a proper datetime format to ensure consistency. All column headers were standardized by removing spaces for better readability and uniformity. Additionally, the data types across the dataset were checked and corrected where needed, ensuring that each column holds the appropriate data type for accurate analysis.
